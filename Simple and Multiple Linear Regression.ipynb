# #In this session, I will cover simple and Multiple Regression Models which are the most fundamental Machine Learning Models 

```python
#import libraries
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
df = pd.read_csv('C:\\Users\\hp\\Desktop\\Reviving python\\Datasets\\rain.csv')
df.head()
```

```python
df.describe()
```

```python
df.plot (x = 'MinTemp', y = 'MaxTemp', style = 'o')
plt.title('Maximum vs minimum Temparatures duting WW2 vs Altitude')  
plt.xlabel('Minimum')  
plt.ylabel('Maximum')  
plt.show()
```

check and plot the average maximum temparature

```python
plt.figure(figsize = (15,10))
plt.tight_layout()
seabornInstance.distplot(df['MaxTemp'])
```

The next step is to divede the data into atttributes (independent variables)and labels(dependent variables). Our attribute here is minimum temparature which we will use to predict the dependent variable i.e maximum temparature
we will store the attribute as X and label as Y

```python
X = df['MinTemp'].values.reshape(-1,1)
y = df['MaxTemp'].values.reshape(-1,1)
```

Next we split our data into the training and test data sets (80% vs 20%)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

After splitting the data sets, the nest step is to train the alogrithim. 

# Linear Regression

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train) #this line of code trains the alogrithm
```

We finally extract the slope and Y intercept. Remember this what the linear ref=gression does, it selects the line of best fit and describes it using the slope and the Y intercept

```python
print(regressor.intercept_) # gives Y intercept of line of best fit
print(regressor.coef_) #gives the slope of the line of best fit
```

this means for every unit change in minimum temparature theres a 0.92% change in the maximum temparature

Since we have successfuly trained the model, the next step is to make predictions. Lets see how accurately our alogrithm predicts the percentage score.  We will use our test data

```python
y_pred = regressor.predict (X_test)
```

Next step is to compare the actual data with the predicted data using the line of code below

```python
data = pd.DataFrame({'Actual':y_test.flatten(),'Predicted': y_pred.flatten()})
data
```

lets plot it as a bargraph for the visuals using the first 25 entries


```python
data25 = data.head(25)
data25.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

next we plot the straigght line using the test data to check the accuracy

```python
plt.scatter(X_test, y_test, color = 'green')
plt.plot(X_test, y_pred, color = 'red', linewidth = 2)
plt.show()
```

the final step entails the evaluation of the alogrithm. This step is key as it compares the performance of different alogrithms of a certain dataset. In Regression alogrithms, three evaluation matrices are used including:
1. Mean Absolute Error
2. Mean Squared error
3. Root Mean Squared Error 
This is done using functions in the SciKit Library


```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

The RMSE is 4.19 which is more than 10% of the mean temparature (22.4). This shows that our model is not very accurate but it demonstrates how to carry out linear regressions 😜

# Multiple Linear Regression

The linear regression above involved only two variables but lets be honest in real life, multiple factors come into play to give one output. We will use multiple linear regression models to determine which factors have greater impact on the output. For this activity we will use a data set related to the red variants of the Portuguese “Vinho Verde” wine. This data set contains input variables like fixed acidity, volatile acidity, citric acid, residual sugar, chlorides, free sulfur dioxide, total sulfur dioxide, density, pH, sulphates, alcohol which determine the overall quality of the wine. 

```python
wine = pd.read_csv('C:\\Users\\hp\\Desktop\\Reviving python\\Datasets\\wine.csv')
```

```python
wine.head()
```

```python
wine.shape
```

```python
wine.describe()
```

```python
wine.isnull().any()
```

Awsome!!no missing values. Incase there were missing values, we would excecute the following line of code to remove any null values.

wine= wine.fillna(method='ffill')

#we split the dataset into Attributes and Labels 
#Remember attribues are the independent variables and Labels are the dependent variables. 

```python
X = wine [['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 
        'total sulfur dioxide', 'density', 'pH', 'sulphates','alcohol']]. values

y = wine['quality'].values
```

```python
#lets have a look at quality
plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(wine['quality'])
```

from the graph its clear that most times the wine quality is 5/6 which is probably the preferred one 

SPLIT THE DATA INTO TRAINING AND TEST DATA 


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

TRAIN THE MODEL

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train) #this line of code trains the alogrithm
```

next we want to find out the coefficient that the model has found for all the attributes

```python
coeff_df = pd.DataFrame({'Coefficient': regressor.coef_})
coeff_df.index = [['fixed acidity', 'volatile acidity', 'citric acid', 
                   'residual sugar', 'chlorides', 'free sulfur dioxide', 
                   'total sulfur dioxide', 'density', 'pH', 'sulphates','alcohol']]
coeff_df
```

To explain this: Density has the highest value of 31 which means that a one unit increase in density, leads to a reduction of wine quality by 31 units. Chlorides and volatile acidity have a bearing on the wine qulity while the rest do not. 
Therefore, DENSITY is the main determinant of wine quality. 

```python
#Now that we have trained our model, lets test it using the test data
y_pred = regressor.predict(X_test)
```

```python
#lets compare the actual and the predicted values
wine2 = pd.DataFrame({'Actual':y_test.flatten(),'Predicted': y_pred.flatten()})
wine2
```

```python
#lets plot out the first 25 entries to visulaize
wine2 = wine2.head(25)
wine2.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

From the look of it, the model has given a pretty good prediction! 

The next step is to evaluate our model using the three variables.
1. Mean Absolute Error
2. Mean Squared error
3. Root Mean Squared Error 

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```


We see that the value of root mean squared error is 0.62, which is slightly greater than 10% of the mean value which is 5.63. This means that our algorithm was not very accurate but can still make reasonably good predictions.

There are many factors that may have contributed to this inaccuracy, for example :

 1. Need more data: We need to have a huge amount of data to get the best possible prediction.

 2. Bad assumptions: We made the assumption that this data has a linear relationship, but that might not be the case. Visualizing the data may help you determine that.

 3. Poor features: The features we used may not have had a high enough correlation to the values we were trying to predict.
